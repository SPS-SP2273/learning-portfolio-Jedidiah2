## Exercise 1 :  F-strings to the rescue

```python
count=None
for i in range(10):
    count=f'{i+1}'
    for length in range(5-len(count)):
        count='0'+count
    print(count)
```

```python
count=None
for i in range(10):
    count=f'{i+1}'
    for length in range(3-len(count)):
        count='0'+count
    count=count+' '*(10-i)+'+'*(i+1)
    print(count)
```

```python
count=None
for i in range(10):
    count=f'{i+1}'
    for length in range(3-len(count)):
        count='0'+count
    count=count+' '+'-'*(1+i)+'+'*(9-i)
    print(count)
```

```python
hero=['bruce','natasha','tony']
info=['exploits','enemies','friends']
files=[]

for h in hero:
    for i in info:
        files.append(f'{h}_{i}.txt')

print(files)
```

```python
country=['sg','gb','us']
experiment=['exp-01','exp-02','exp-03']
files=[]

for c in country:
    for e in experiment:
        for x in range(50):
            count=f'{x+1}'
            for length in range(3-len(count)):
                count='0'+count
            files.append(f'data_{c}_{e}_trial-{count}.txt')
        
print(files)
```

## Exercise 4 :  A function for Fibonacci

```python
def fibonnaci(n):
    try:
        fibonnaci=[0,1]
        if n < 0:
            print('Please input a positive integer')
        elif 0 <= n < 2:
            return fibonnaci[:n]
        else:
            for i in range(n-2):
                reverse=fibonnaci[::-1]
                fibonnaci.append(reverse[0]+reverse[1])
            return fibonnaci
    except:
        print('Please input integer \'n\' for first n values of the Fibonacci sequence')

fibonnaci(7)
```

## Exercise 5 :  max_info() without NumPy

```python
def max_info(numbers):
    my_max = numbers[0]
    for i, n in enumerate(numbers):
        if n > my_max:
            my_max = n
            my_max_index = i
    return my_max, my_max_index

test = [40, 27, 83, 44, 74, 51, 76, 77, 10, 49]
max_info(test)
```

```python
def ACGT(genome):
    A = []
    C = []
    G = []
    T = []
    for i in genome:
        if i == 'A':
            A.append('A')
        elif i == 'C':
            C.append('C')
        elif i == 'G':
            G.append('G')
        elif i == 'T':
            T.append('T')
    return A, C, G, T

ACGT(DNA)
```
