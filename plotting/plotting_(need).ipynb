## 1 A simple plot

### 1.1 Letâ€™s look at some code.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting (https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html)
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=1, label='Y values')   # color (c), linestyle (ls), linewidth (lw)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.8)               # gridline darkness (e.g. .25 is lighter than .8)
plt.legend(loc='lower right')    # where legend is: upper/lower, left/right; center; best
plt.show()                       # good practice to run this line (graph will show in jupyter but not in all environments)
```

```python
plt.plot(x, y_2)
```

### 1.2 Adding another plot

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')    # to add another line
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
plt.show()
```

### 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .5, 0.75    # can adjust; error shown is according to axis squares

# can also provide a different error for each point by placing it in a list

# Lets start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(x, y_2, color='blue', label='Y2 values')
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             color='green', label='Y3 with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='best')
plt.show()
```

### 1.5 Better with NumPy

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.show()
```

### 1.6 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

### 1.7 We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

```python
x = np.linspace(0, 6, 100)  #Return evenly spaced numbers over a specified interval
# np.linespace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
# num: number of points to generate

x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)   # to save in current directory
# filetypes= .png, .jpeg, .pdf, .bmp; dpi = resolution (higher is better)

# plt.savefig('C://Desktop/simple-plot.png', dpi=150) to save in specified directory
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
plt.style.use('fivethirtyeight')   # to change the style of the plot

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)   # plot(x, y)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')   # horizontal line (hlines(y value, starting x, ending x,))
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```

## Exercise 1 :  Fill in the blanks

```python
plt.style.use('default')
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, xerr=x_err, yerr=y_err,
             color='#99000d', linestyle='--',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show(block=False)
```

## Exercise 2 :  A simple mathematical plot

```python
x = np.linspace(0, 50, 500)
y = np.exp(-x/10)*np.sin(x)
# print(x)
# print(y)
plt.plot(x,y)
plt.xlabel('x', fontsize = 15, color = 'darkblue')
plt.ylabel('y', fontsize = 15, color = 'darkblue', rotation = 'horizontal')
plt.grid(alpha = .25)
plt.title('$y=e^{-x/10}\sin(x)$', fontsize = 20)
plt.ylim([-1,1])
plt.show()
```
